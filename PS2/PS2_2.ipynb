```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("2281305.csv")
```

```python
df['WOda'] = df['WND'].apply(lambda x: x.split(',')[0])  
df['WOdaqc'] = df['WND'].apply(lambda x: x.split(',')[1]) 
df['WOtc'] = df['WND'].apply(lambda x: x.split(',')[2]) 
df['WOsr'] = df['WND'].apply(lambda x: x.split(',')[3]).astype(int)
df['WOsqc'] = df['WND'].apply(lambda x: x.split(',')[4]).astype(int)
df['Year-Month'] = df['DATE'].apply(lambda x: x.split('-')[0]+x.split('-')[1]) 
df_count = df.loc[df['WOsqc'] < 2].groupby('Year-Month')['WOsr'].count()
df_speed = df.loc[df['WOsqc'] < 2].groupby('Year-Month')['WOsr'].sum()
df_mean = (df_speed / df_count) * 0.1
df_mean = pd.DataFrame(df_mean)
fig=plt.figure(num=1,figsize=(9,4))
plt.plot(df_mean[:12])
plt.show()
fig=plt.figure(num=2,figsize=(9,4))
plt.plot(df_mean[12:24])
plt.show()
fig=plt.figure(num=3,figsize=(9,4))
plt.plot(df_mean[24:36])
plt.show()
fig=plt.figure(num=4,figsize=(9,4))
plt.plot(df_mean[36:48])
plt.show()
fig=plt.figure(num=5,figsize=(9,4))
plt.plot(df_mean[48:60])
plt.show()
fig=plt.figure(num=6,figsize=(9,4))
plt.plot(df_mean[60:72])
plt.show()
fig=plt.figure(num=7,figsize=(9,4))
plt.plot(df_mean[72:84])
plt.show()
fig=plt.figure(num=8,figsize=(9,4))
plt.plot(df_mean[84:96])
plt.show()
fig=plt.figure(num=9,figsize=(9,4))
plt.plot(df_mean[96:108])
plt.show()
fig=plt.figure(num=10,figsize=(9,4))
plt.plot(df_mean[108:120])
plt.show()
fig=plt.figure(num=11,figsize=(9,4))
plt.plot(df_mean[120:])
plt.show()
```
